```python
# Leitura dos dados de entrada
# Como os valores vêm em linhas separadas, lemos um de cada vez
e = int(input())
d = int(input())

# Aplicação das regras da brincadeira
if e > d:
    # Se a mão esquerda for maior, o resultado é a soma
    resultado = e + d
else:
    # Caso contrário (E < D), o resultado é o dobro da diferença
    resultado = 2 * (d - e)

# Impressão do resultado final
print(resultado)
```

```python
# Leitura da entrada (cada valor em uma linha)
h = int(input())
m = int(input())
s = int(input())
t = int(input())

# 1. Converter o horário atual para o total de segundos desde o início do dia
# Um minuto tem 60 segundos; uma hora tem 3600 segundos (60 * 60)
total_segundos = h * 3600 + m * 60 + s

# 2. Somar o tempo de adiamento
total_segundos += t

# 3. Como o tempo pode passar de um dia, usamos o total de segundos em 24h
# 24 horas * 3600 segundos = 86.400 segundos
total_segundos = total_segundos % 86400

# 4. Converter de volta para H, M, S
# Extraímos as horas dividindo o total por 3600
nova_h = total_segundos // 3600
# O que sobra (resto) são os minutos e segundos
resto_segundos = total_segundos % 3600

# Dos segundos restantes, dividimos por 60 para achar os minutos
nova_m = resto_segundos // 60
# O que sobrar dessa divisão são os segundos finais
nova_s = resto_segundos % 60

# Impressão seguindo o formato solicitado
print(nova_h)
print(nova_m)
print(nova_s)
```

```python
# 1. Lemos a linha com N e K como uma string e dividimos nos espaços
entrada = input().split()

# 2. Pegamos cada valor da lista 'entrada' e convertemos para inteiro manualmente
n = int(entrada[0])
k = int(entrada[1])

# 3. Lemos a linha das notas e dividimos nos espaços
notas_texto = input().split()

# 4. Criamos uma lista vazia para guardar as notas como números
notas = []

# 5. Percorremos a lista de textos, transformamos cada um em número e guardamos
for item in notas_texto:
    nota_convertida = int(item)
    notas.append(nota_convertida)

# 6. Ordenamos do maior para o menor
notas.sort(reverse=True)

# 7. A nota de corte é o elemento na posição K
# (Lembrando que em computação começamos a contar do 0)
print(notas[k-1])
```

```python
# 1. Leitura de N (tamanho) e Q (passos)
entrada = input().split()
n = int(entrada[0])
q = int(entrada[1])

# 2. Leitura da matriz inicial
matriz = []
for i in range(n):
    linha_texto = input()
    linha_numeros = []
    for caractere in linha_texto:
        linha_numeros.append(int(caractere))
    matriz.append(linha_numeros)

# 3. Simulação por Q passos
for passo in range(q):
    # Criamos uma nova matriz vazia para o próximo estado
    proxima_matriz = []
    for i in range(n):
        proxima_matriz.append([0] * n)

    # Para cada célula da matriz...
    for r in range(n):
        for c in range(n):
            # Contamos os vizinhos vivos (das 8 direções)
            vizinhos_vivos = 0

            # Checamos as células ao redor (de -1 a +1 em linha e coluna)
            for dr in [-1, 0, 1]:
                for dc in [-1, 0, 1]:
                    if dr == 0 and dc == 0:
                        continue # Pula a própria célula

                    nr, nc = r + dr, c + dc # Nova linha e nova coluna

                    # Verifica se o vizinho está dentro dos limites da matriz
                    if 0 <= nr < n and 0 <= nc < n:
                        if matriz[nr][nc] == 1:
                            vizinhos_vivos += 1

            # Aplicamos as regras de Conway
            if matriz[r][c] == 1: # Célula está VIVA
                if vizinhos_vivos == 2 or vizinhos_vivos == 3:
                    proxima_matriz[r][c] = 1 # Continua viva
                else:
                    proxima_matriz[r][c] = 0 # Morre
            else: # Célula está MORTA
                if vizinhos_vivos == 3:
                    proxima_matriz[r][c] = 1 # Nasce
                else:
                    proxima_matriz[r][c] = 0 # Continua morta

    # Após calcular tudo, a matriz atual vira a próxima
    matriz = proxima_matriz

# 4. Impressão do resultado final
for linha in matriz:
    # Transformamos os números de volta em texto para imprimir sem espaços
    resultado_linha = ""
    for num in linha:
        resultado_linha += str(num)
    print(resultado_linha)
```
