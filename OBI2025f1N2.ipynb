```python
# Lê as posições (lembre-se: sem texto no input para a OBI)
E = int(input())
S = int(input())
L = int(input())

# Começamos assumindo que a escola (E) é o maior e o menor
maior = E
menor = E

# Testamos o supermercado (S)
if S > maior:
    maior = S
if S < menor:
    menor = S

# Testamos a lojinha (L)
if L > maior:
    maior = L
if L < menor:
    menor = L

# A distância total é o dobro da diferença entre os extremos
distancia = 2 * (maior - menor)

print(distancia)
```

```python
# Lê a quantidade de refeições (N) e o limite (M)
# O split() serve para separar os dois números da primeira linha
dados = input().split()
N = int(dados[0])
M = int(dados[1])

total_consumido = 0

# Repete o processo para cada uma das N refeições
for i in range(N):
    # Lê as gramas de Proteína, Gordura e Carboidrato
    refeicao = input().split()
    P = int(refeicao[0])
    G = int(refeicao[1])
    C = int(refeicao[2])

    # Calcula as calorias desta refeição e soma ao total
    # 1g Proteína = 4 cal, 1g Gordura = 9 cal, 1g Carboidrato = 4 cal
    calorias = (P * 4) + (G * 9) + (C * 4)
    total_consumido = total_consumido + calorias

# A resposta é o limite menos o que ele já comeu
print(M - total_consumido)
```

```python
# Lê os quatro valores (um em cada linha, então não precisa de split)
A = int(input())
B = int(input())
C = int(input())
D = int(input())

# O café necessário deve estar entre esses dois valores:
cafe_minimo = C - B
cafe_maximo = C - A

# Criamos uma variável para guardar se deu certo (começa como "Não")
possivel = "N"

# Testamos de 1 dose até 50 doses (50 é mais que o suficiente para 500ml)
for doses in range(1, 51):
    total_cafe = doses * D

    # Se o café dessa dose estiver no intervalo desejado, deu certo!
    if total_cafe >= cafe_minimo and total_cafe <= cafe_maximo:
        possivel = "S"

# Imprime o resultado final (S ou N)
print(possivel)
```

```python
# 1. Lê a quantidade de brinquedos
N = int(input())

# 2. Lê a linha toda como um texto e "corta" nos espaços
lista_texto = input().split()

# 3. Criamos uma lista vazia para guardar os números de verdade
X = []

# 4. Para cada pedaço de texto, transformamos em número e guardamos na lista X
for pedaço in lista_texto:
    numero = int(pedaço)
    X.append(numero)

# 5. Descobrimos qual é o maior número para saber a altura do gráfico
maior = 0
for valor in X:
    if valor > maior:
        maior = valor

# 6. Começamos a desenhar do andar mais alto (maior) até o chão (1)
andar = maior
while andar > 0:

    # Para cada brinquedo, verificamos se ele alcança esse andar
    for valor in X:
        if valor >= andar:
            print(1, end=" ")
        else:
            print(0, end=" ")

    # Pulamos para a linha de baixo e descemos um andar
    print()
    andar = andar - 1
```
