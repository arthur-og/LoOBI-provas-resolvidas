### 1. Pizzaria (`pizzaria.py`)
A mãe de Larissa decidiu comemorar o aniversário da filha em uma pizzaria.
* **Pizza Grande:** 8 fatias.
* **Pizza Pequena:** 4 fatias.

Larissa quer saber o número máximo de amigos que pode convidar, garantindo uma fatia para ela, uma para a mãe e uma para cada amigo. O cálculo matemático é:
$$\text{Amigos} = (G \times 8 + P \times 4) - 2$$

**Entrada:** Inteiros $G$ e $P$ ($1 \le G \le 10$ e $1 \le P \le 10$).
**Saída:** Número máximo de amigos.

| Entrada | Saída |
| :--- | :--- |
| 2<br>3 | 26 |
| 7<br>4 | 70 |
| 1<br>2 | 14 |

---



```python
G = int(input())
P = int(input())
Amigos = (G * 8 + P * 4) - 2
print(Amigos)
```

### 2. Café com Leite (`leite.py`)
O cliente indica a capacidade $C$ da xícara e o volume de leite desejado, entre $A$ (mínimo) e $B$ (máximo). A máquina prepara exatamente $D$ mililitros de café. O volume de leite adicionado para encher a xícara será $C - D$.
O pedido será satisfeito se a seguinte condição for verdadeira:
$$A \le (C - D) \le B$$

**Entrada:** Quatro inteiros: $A$, $B$, $C$, $D$ ($100 < C < 500$, $0 \le A < B < C$, $10 < D < 100$).
**Saída:** `S` se o volume estiver correto, `N` caso contrário.

| Entrada | Saída |
| :--- | :--- |
| 130<br>150<br>170<br>30 | S |
| 220<br>230<br>240<br>30 | N |
| 0<br>200<br>300<br>45 | N |

---



```python
A = int(input())
B = int(input())
C = int(input())
D = int(input())
print("S" if A <= (C - D) <= B else "N")
```

### 3. Fila (`fila.py`)
O professor Ricardo vigia $N$ alunos. O aluno na cadeira $i$ tem altura $a_i$. A cadeira 1 fica no fundo e a cadeira $N$ na frente. Ricardo não enxerga o aluno na cadeira $i$ se existir um aluno na cadeira $j$ tal que:
$$j > i \quad \text{e} \quad a_j \ge a_i$$

**Entrada:**
* Inteiro $N$ ($1 \le N \le 100.000$).
* $N$ inteiros $a_1, a_2, \dots, a_N$ ($1 \le a_i \le 1.000.000$).

**Saída:** Quantidade de alunos que Ricardo não consegue enxergar.

| Entrada | Saída |
| :--- | :--- |
| 5<br>200 180 190 140 160 | 2 |
| 5<br>180 180 170 150 190 | 4 |
| 2<br>150 150 | 1 |

```python
n = int(input())
alturas = [int(x) for x in input().split()]
escondidos = 0
maior_frente = 0

for altura in reversed(alturas):
    if altura <= maior_frente:
        escondidos += 1
    else:
        maior_frente = altura

print(escondidos)
```
